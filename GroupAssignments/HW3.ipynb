```python
#dtype NRI_ID  and statefips ensures that the columns are read in as strings for leading zeros
ni = pd.read_csv(
    'data/National_Risk_Index_Census_Tracts.csv',
    dtype=
    {
        'NRI_ID':str,
        'STATEFIPS': str
    }
)
```

```python
# show top 5 rows of a specific columns
ni.STATEFIPS.head()
```

```python
# I only want the data that was in CA
new = ni[ni['STATEABBRV'] == 'CA']
```

```python
new.head(3)
```

I want Geo_FIps so I want to delete T0 in NRI_ID

```python
new['NRI_ID'] = new['NRI_ID'].str.replace('T', '')

#new.drop(1, inplace=True) 
```

```python
#it is int, so 0 cannot show, I want transform to transform int to string
new['NRI_ID'] = new['NRI_ID'].astype(str)
```

```python
new.NRI_ID.head()
```

Now we can see the 0 before 6

```python
#new.Geo_FIPS=new.NRI_ID  Change the original name to GEO_id
new = new.rename(columns={'GEO_id': 'FIPS'})
```

```python
new.head()
```

```python
#find the columns that comtains null value
new.columns[new.isna().all()].tolist()
```

I want to delete those empty columns to make sure the survey is correct

```python
# axis = 0 delete rows and axis = 1 delete columns with null
new = new.dropna(axis=1,how="all")
```
